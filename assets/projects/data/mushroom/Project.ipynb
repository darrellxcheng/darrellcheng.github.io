 Dataset Information: 
 >From Audobon Society Field Guide; mushrooms described in terms of physical characteristics; classification: poisonous or edible

## Basic Data Visualization 

Basic visualizations of the dataset using different Seaborn plot options, to better understand patterns within the data.

```python
# Joining the two dataframes 
mushrooms = pd.concat([mushroomx,mushroomy], axis = 1)
mushrooms
```

```python
sns.countplot(x='poisonous', hue='poisonous', data=mushrooms, palette="Set1", width = 0.5)
plt.title ('Distribution of Poisonous Measures as Direct Count')
plt.show()
```

### Cap Features

```python
cap_features = ['cap-shape', 'cap-surface', 'cap-color'] #Simple array of cap features

fig, axes = plt.subplots(1, 3, figsize=(18, 4)) #Divide the plot into a 1x3 matrix of subplots with plot size 18,4

for graph, cap in zip(axes, cap_features): #Loops through the features and attaches the iterable objects together using Python's zip() method
    sns.countplot(data=mushrooms, hue=cap, x='poisonous', ax=graph) #Seaborn countplot utilizing 'mushrooms' data and the 'graph' variable set in the loop.
    graph.set_xlabel('Edibility')
    graph.set_title(f'Distribution of {cap}')
    graph.legend(title=cap, loc='upper right') #These last three should be pretty self explanatory
#Visual analysis of the information via Seaborn

plt.tight_layout() #Adjusts the padding in and around subplots, added for visual clarity.'

plt.show()
```

From these three graphs, generated from each of the cap-related features of the dataset, we can extrapolate the following (for now):
- Cap shape does not seem to have an extremely drastic effect on the edibility of the sample, save for 'b' (a bell shaped cap) possessing a significantly higher concentration of edible samples as opposed to poisonous ones. There also seem to be some null values or shapes with unknown edibility, which are a result of either missing values from the original dataset, or a lack of testing during collection.
- The surface of the cap seems to have some moderate indications of edibility, although the values are similar in such a way that the determination of ediblity is unlikely to be strongly influenced by any particular sub-feature. That being said, the edibility of groove capped mushrooms seems to be indeterminable, and it is far more likely for fibrous capped mushrooms to be edible.
- Cap color seems to be the most consistent in its poison/edibility split, with minor nuances in the two targets, save for the likelihood of white-capped species being statistically mroe edible. 

Considering these factors, the features of the mushroom's cap are not an extremely accurate method in determining edibility, especially when many samples based purely off these features have an unknown edibility; or, since the concentrations of each sub-feature is roughly similar in spite of the comparable amounts of surely poisonous/surely edible mushrooms.

Here we see our first reasonably visible distinction within the data. ### Let's find out the exact number below:

```python
non_smelling = mushrooms[mushrooms['odor'] == 'n'] #Create new datafram that is a representation of all values in column 'odor' with the value 'n' (none)
ns_count = non_smelling['poisonous'].value_counts() #Count the values of column 'poisonous' in non smelling dataframe
print(ns_count) #Print Value
total_ns = ns_count.sum() #Calculate total amount of non smelling samples
ns_pe = (ns_count['e'] / total_ns) #Use total count of non smelling samples to find amount of 'e' edible samples, and divide this by the total to find the percentage.
print (ns_pe * 100) #Print Value

```

We see that there are 3,288 more edible samples of non-smelling mushrooms than there are of poisonous ones and that if a species has no odor, it is ~96.599% more likely to be edible.

4. Gill color seems to be the most visually complex graph, although this is likely due to the larger array of available data (more colors). The data here is more varied, with several deviations of varying degrees. For instance, black mushrooms are more likely to be edible, and brown mushrooms are far more likely to be edible than poisonous. Gray and 'chocolate' colored samples are more likely to be poisonous, and 'buff' colores samples are almost certainly poisonous in every instance. Pink, white, and purple mushrooms are also more likely to be edible than poisonous by a reasonable amount, while other colors are either missing data, of an unknown edibility, or the sample counts are closely aligned such that edibility is an even (or close to) split.

## Advanced Data Visualization

In this section, we'll see how different selections of traits impact the edibility of a sample within the dataset. ### Stacked Bar Plots

Stacked bar plots will assist in a simple comparison in showing the frequency of features falling within each category, and also display the distribution of edibility in relation to any combination of specific traits.

```python
ct_cap = pd.crosstab(index=[mushrooms['cap-shape'], mushrooms['cap-surface'], mushrooms['cap-color']], columns=mushrooms['poisonous']) #Crosstab of the cap features, a cross tabulation of two or more objects, see documentation here: https://pandas.pydata.org/docs/reference/api/pandas.crosstab.html. Computes a frequency table of the factors.
ct_cap.plot(kind='bar', stacked=True, figsize=(20, 6)) #Predetermine the plot features (kind, stacked as well as the graph size)
plt.title('Stacked Bar Chart of Cap Features by Edibility')
plt.xlabel('Cap Shape, Cap Surface, Cap Color')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
#The rest should all be reasonably self-explanatory
```

We can see from the stacked bar plot that there are quite a few combinations of cap features that may affect mushroom edibility. Each combination is marked on the x-axis of the chart, and the edibility of each of those combinations is marked as a stacked bar graph. However, it seems a little difficult to really extrapolate extensive information from this method. We can see that combinations such as (x,f,e) and (x,y,g) may have higher representation compared to other combinations and thus have more varied edibility, but let's find a way to learn a little bit more.

```python
ct_cap['Total'] = ct_cap.sum(axis=1) #Add the counts of each combination in the crosstabulation and use them to create a new column in the table displaying totals
ct_cap = ct_cap.sort_values(by='e', ascending=False) #Sort the values in the crosstab table in descending order by edibility
ct_cap_p = ct_cap.sort_values(by='p', ascending=False)#Same as above but inverse (sort by highest distribution of poisonous first)
ct_cap.plot(kind='bar', stacked=True, figsize=(20, 6))#Plot the table as a stacked bar plot

sorted_ct_cap = ct_cap.reset_index() #Reset the index values of the DataFrame, using the default one instead in order to normalize the values 
sorted_ct_cap = sorted_ct_cap.sort_values(by='Total', ascending=False)#Sort the reset dataframe by the totals to see whole distributions of the combinations in descending order
sorted_ct_cap.plot(kind='bar', stacked=True, figsize=(20, 6))#Plot the sorted datafram as a bar plot
plt.xticks(rotation=45)

print(sorted_ct_cap.head()) 
print(ct_cap.head(10))
print(ct_cap_p.head(10)) #Print the first few entries of each of the tables
```

So, from the new graphs and sorted data, we can now make some more concrete observations.
1. When it comes to cap features, the combination of (x,f,g) is the most prevalent within the data, representing 442 samples. This combination includes a convex cap shape, a fibrous surface, and a gray color. (x,y,n) and (f,y,n) represent the second most common distribution at 390 samples each. These samples possess a convex or flat shape, scaly surface, and brown color. 
2. We can also derive much more important information from these graphs. For instance, when it comes to edibility, combination (x,f,g) has the most edibility at 248 samples, but also a similar amount of poisonous ones at 194, which makes its true edibility more of a cointoss. However, (x,f,n) and (f,f,n) has 228 edible samples with only 2 poisonous ones, making these two combinations highly likely to be edible. Therefore, a mushroom with cap features of a convex cap or flat cap, fibrous surface, and brown color are highly likely to be edible. We can also observe (b,s,w), (x,s,y), (f,f,e), (x,f,e), (f,f,w), (b,y,w), and (b,y,w) possess similar ratios of edibility.
3. From the inverse table, we can see that (x,f,g) also possesses the highest number of poisonous samples, but the ratios remain similar for the first few entries when comparing totals. From the graph we can osberve that (k,s,n) and (f,y,y) have the highest ratio of poisonous samples to total amount. So if a mushroom's cap is knobbed, smooth, and brown or flat, scaly, and yellow, they are reasonably more likely to be poisonous than edible.
4. Additionally, samples (k,y,n), (x,f,y) and (f,f,y) possess extremly high ratios of poisonous sample distribution to edible. Thus, a mushroom that is (knobbed, scaly, brown), (convex, fibrous, yellow), or (flat, fibrous, yellow) are exceptionally likely to be poisonous.

## Data Prediction

While we saw visualizations of high importance features in the previous section, let's get a more accurate representation using some machine learning methods in order to gather better visualizations and direct numeric counts.

```python
print(oh_corr['poisonous_p'].sort_values(ascending=False).head(10))
print(le_corr['poisonous'].sort_values(ascending=False).head(10))
```

We see from the printed graphs of the correlation matrices that certain traits contribute heavily to a sample being poisonous compared to others: the specific features are only considered in the one-hit encoded matrix due to increased dimensionality however, so keep this in mind! It seems like odor, stalk surface above and below the ring, gill size, bruises, and spore print color seem to have the heaviest effects. Let's continue and make it more accurate and also generate a visualization using machine learning.

```python
#Importing machine learning libraries
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
#Defining input and output
le_x = m_le.drop('poisonous', axis=1)
le_y = m_le['poisonous']
#Splitting dataset into training and testing data at 80/20% (Pareto Principle)
X_train, X_test, y_train, y_test = train_test_split(le_x, le_y, test_size = 0.2, random_state = 5)
RF = RandomForestClassifier()
#Fitting scikit learn classifier to data
RF.fit(X_train, y_train)

feature_importance = pd.DataFrame(
{'feature':list(X_train.columns),
'feature_importance':[abs(i) for i in RF.feature_importances_]}
)
#Create dataframe to display feature importance as a table and use absolute values for the right hand values of the table.
feature_importance_sorted = feature_importance.sort_values(by='feature_importance', ascending=False) #Sort the feature importance table

print(feature_importance_sorted.to_markdown()) #Print sorted table

```

```python
#Defining input and output
oh_x = m_1h.drop(['poisonous_e','poisonous_p'], axis=1)
oh_y = m_1h['poisonous_p']
#Splitting dataset into training and testing data at 80/20% (Pareto Principle)
X_train, X_test, y_train, y_test = train_test_split(oh_x, oh_y, test_size = 0.2, random_state = 5)
RF = RandomForestClassifier()
#Fitting scikit learn classifier to data
RF.fit(X_train, y_train)

feature_importance = pd.DataFrame(
{'feature':list(X_train.columns),
'feature_importance':[abs(i) for i in RF.feature_importances_]}
)
#Create dataframe to display feature importance as a table and use absolute values for the right hand values of the table.
feature_importance_sorted = feature_importance.sort_values(by='feature_importance', ascending=False) #Sort the feature importance table

print(feature_importance_sorted.to_markdown()) #Print sorted table

```

Let's visualize this data now with a very similar approach.

```python
#Defining input and output
le_x = m_le.drop('poisonous', axis=1)
le_y = m_le['poisonous']
#Splitting dataset into training and testing data at 80/20% (Pareto Principle)
X_train, X_test, y_train, y_test = train_test_split(le_x, le_y, test_size = 0.2, random_state = 5)
RF = RandomForestClassifier(random_state=5)
#Fitting scikit learn classifier to data
RF.fit(X_train, y_train)
# Get feature importances
importances = RF.feature_importances_
indices = np.argsort(importances)[::-1] #Sort data in descending order
feature_names = [X_train.columns[i] for i in indices] #Assign names from the indices loop
# Create plot
plt.figure(figsize=(12, 8))
plt.title("Feature Importance")
plt.barh(range(len(indices)), importances[indices], align= 'center')
plt.gca().invert_yaxis()
# Add feature names as y-axis labels
plt.yticks(range(len(indices)), feature_names)
plt.xlabel('Importance')
plt.tight_layout()
plt.show()
```

As we can see from the visualization and tables, it seems as if our predictions were correct. From this, we can extrapolate some data, namely the factors that have the greatest chance at determining a mushroom's edibility:
1. If the mushroom possesses an odor at all, *especially* an unpleasant or foul one, it is very likely to be poisonous.
2. A mushrooms with narrow gills is likely to be poisonous, as well as ones where the gills are buff, chocolate, or gray colored.
3. A mushroom with a spore print that is white, red, or chocolate are likely to be poisonous.
4. A mushroom with bruises is much more likely to be edible than poisonous!
5. A mushroom with a silky stalk is significantly more likely to be poisonous than edible.

These combinations of traits seem to contribute most to edibility. Ergo, through examining this data and its distribution, we can come to generate ideas of what kinds of mushrooms to watch out for when foraging in the wild.
